```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
from sqlalchemy import create_engine
```

```python
# import csvs
statedata_csv = "2015_statedata.csv"
statedata_df = pd.read_csv(statedata_csv)
incidence_csv = "incidence.csv"
incidence_df = pd.read_csv(incidence_csv)
```

```python
# save dataframes to sqlite database
engine = create_engine('sqlite:///statecancerdata.sqlite', echo=False)
statedata_df.to_sql('statedata', con=engine, if_exists='replace', index=False)
incidence_df.to_sql('incidence', con=engine, if_exists='replace', index=False)
```

```python
pd.read_sql_query("SELECT * FROM statedata", con=engine).head()
```
